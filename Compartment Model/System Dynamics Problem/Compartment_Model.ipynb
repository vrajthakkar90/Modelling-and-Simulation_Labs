1(a) With a system dynamics tool or a computer program, develop a model for
a radioactive chain of three elements, from substanceA to substanceB to
substanceC. Allow the user to designate constants. Generate a graph and a
table for the amounts of substanceA, substanceB, and substanceC versus
time. Answer the following questions using this model.

```python
# for a
# let us assume that the initial values for substance A = 10 and B=C=0
A=10
B=0
C=0
# define the time interval as dt=0.1s
dt=0.1
# now define the rate for A B C (random values provided by user)
a=0.2
b=0.3

it=0
QuantA=[A]
QuantB=[B]
QuantC=[C]
iter=[0]
t=0
plt.show()
while(t<20):
  t+=dt
  A = A - a*A*(dt)
  QuantA.append(A)
  B = B + a*A*dt - b*B*dt
  QuantB.append(B)
  C= C + b*B*dt
  QuantC.append(C)
  iter.append(t)
  it+=1

plt.plot(iter,QuantA, label = 'Substance A')
plt.plot(iter,QuantB, color='r', label = 'Substance B')
plt.plot(iter,QuantC, color= 'g', label = 'Substance C')
plt.legend()
plt.xlabel('Time_stamp')
plt.ylabel('Quantity')

```

(C) *As the decay rate of A, a, increases from 0.1 to 1, describe how the time of
the maximum total radioactivity changes. The total radioactivity is the
sum of the change from substanceA to substanceB and the change from
substanceB to substanceC, or the total number of disintegrations. Why?*

Answer:
* The time of maximum total radioactivity changes is the time instant when the number of particle emissions in the system are maximum. That is when (a*A + b*B)*dt is maximum. By plotting the total radioactivity with time_stamp we can determine the time_stamp at which maximum total radioactivity occurs. Also noticing that the value depends on the rate constant of B(b) as well, but only considering a change in the value of rate constant(a) of A and fixing the value of b to a constant value. We notice that maximum total radioactivity occurs at

  1. time_stamp=3.0s for a=0.1 and b=0.3 \\
  2. time_stamp=0.0s for a=1 and b=0.3

* The total radioactivity is the total number of disintegration by definition and so is true for this example. It will not be the sum of change from substanceA to substanceB and the change from substanceB to substanceC because\\

1. change from substanceA to substanceB = change in quantity of substanceB = (a*A - b*B)*dt
2. change from substanceB to substanceC = b*B*dt

Their sum will be a*A*dt which is not the total radioactivity. (As we know that total radioactivity is a*A*dt + b*B*dt)


```python
# for a
# let us assume that the initial values for substance A = 10 and B=C=0
A=10
B=0
C=0
# define the time interval as dt=0.1s
dt=0.5
# now define the rate for A B C (random values provided by user)
a=1
b=0.3

it=0
QuantA=[A]
QuantB=[B]
QuantC=[C]
TR= (a*A + b*B)*(dt)
total_radioactivity=[TR]
iter=[0]
t=0
plt.show()
while(t<20):
  t+=dt
  A = A - a*A*(dt)
  B = B + a*A*dt - b*B*dt
  TR= (a*A + b*B)*(dt)
  total_radioactivity.append(TR)
  iter.append(t)
  it+=1
plt.plot(iter, total_radioactivity)
plt.xlabel('Time_stamp')
plt.ylabel('Total_radioacitivity')

# print the value of time_stamp at which max value occurs

max_index=np.argmax(total_radioactivity)
print(iter[max_index])
```
