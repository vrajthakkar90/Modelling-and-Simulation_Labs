```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#population adapting the product
#p iS assumed to be 0.02
maxpopulation=40000000
population=0
paray=[]
dt=0.1
time=[]
p=0.02
dp=[]
for i in np.arange(0,1000,0.1):
    population= population + p*dt*(maxpopulation-population)
    dn = p*dt*(maxpopulation-population)
    paray.append(population)
    time.append(i)
    dp.append(dn)
plt.plot(time,paray)
plt.xlabel('Time_stamp')
plt.ylabel('Population with information')
```

Here we get an exponential increase like the newton's law of cooling. As the growth rate p which essentially refers to the rate at which information spreads is increased, we see that the population of people with the information reaches to it's maximum quickly which is also very intuitive.

```python
plt.plot(time,dp,color='orange')
plt.xlabel('Time_stamp')
plt.ylabel('Dn(Change in population with information)')
```

The change in the number of people with the information with time gives a exponentially decreasing curve because initially there are a lot of people without information and the population of people gaining the information is directly proportional to C-N(t) where C is total population and N(t) is the amount of people with the information at any given time t. So as N(t) increases C-N(t) decreases and therefore Dn decreases. Dn= N(t+1)-N(t)

```python
#population adapting the product
#p iS assumed to be 0.02
maxpopulation=40000000
population=20
paray=[]
dt=0.1
time=[]
q=0.25
p=0.02
dp=[]
for i in np.arange(0,100,0.1):
    c=population/maxpopulation
    dn= q*dt*(maxpopulation-population)*c
    population= population + q*dt*(maxpopulation-population)*c
    paray.append(population)
    dp.append(dn)
    time.append(i)
plt.plot(time,paray)
plt.xlabel('Time_stamp')
plt.ylabel('Population with information')
```

The plot obtained here is a logistic curve and the growth rate here is q which is the interaction rate and signifies the rate at which people interact to spread the information. Intuitively we would expect that increase in the rate should result in faster spreading of information and faster attainment of the equilibrium or maximum population with the information. Also we must mention that in this model every human is indifferent and has the same properties. Moreover there are no external other factors involved. Also the values of P, Q and total population are taken from Bass's paper. The value of dt or iterative time stamp is taken to be 0.1 to avoid numerical complications.

The above plot signifies the number of people getting the information with their time stamps which a bell shaped curve. This curve signifies that the number of people getting the information per unit time is initially increasing and it increases upto a max value and then decreases until it becomes zero. This can be easily understood from the differential equation that the curve will attain a maxima. But intuitively we see that the number of people who do not have the information will be very large so the information spreads rapidly as more and more people interact but at a point when information has already spread to many people and there are not enough people to spread the information as quickly as before then from that point the curve starts going downwards. From this point the information spread will not be as large because many people already have the information and less people don't have the information. When plotting the value of change in population with the information with respect to total population with information we will find the maximum information will be passed when N(t)=C/2 which can also be verified analytically as follows *idhar analytically dikhana*


```python
#population adapting the product
#p iS assumed to be 0.02
maxpopulation=40000000
population=0
paray=[]
dt=0.1
time=[]
p=0.05
q=0.25
c=1
dp=[]
for i in np.arange(0,100,0.1):
    population+=p*dt*(maxpopulation-population)
    c=population/maxpopulation
    population = population + q*dt*(maxpopulation-population)*c
    paray.append(population)
    dn= q*dt*(maxpopulation-population)*c + p*dt*(maxpopulation-population)
    dp.append(dn)
    time.append(i)
plt.plot(time,paray)
plt.xlabel('Time_stamp')
plt.ylabel('Population with information')
```

The above model is applicable when we have interactions as well as some constant increasing number of people getting the information which an extension to both the problems discussed above. In this model the peak shifts because there are two or we can say increased factors causing the change in the population of the people with information. So the peak which essentially means the time_stamp at which the change in the population of people with information is maximum and after which the change decreases because most of the people who are going to interact already have the information shifts leftwards because there is an increased rate caused due to the summation of both terms. Moreover we also observe that the change in population with information is maximum at some point which is not N(t)= C/2 but occurs before it which also evident from the analytical solution described below.  *Idhar analytical dikha dena*
