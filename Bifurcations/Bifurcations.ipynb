Throughout this lab we explore the idea of bifurcations on various models, particularly we look at local bifurcations. Local bifurcations occur when the stability of an equilibrium point changes between stable and unstable. In this model particularly we obtain a superciritical pitchfork bifurcation.

If the company's market share is above 50 percent and significantly influenced by global market then we need to decrease the strength of influence from the global market by introducing offers, discounts, reducing import exports from the global market. Make changes in the supply system to decrease the global connect will get us more money to focus on our growth and increase our market share.

If the company's market share is below 50 percent than we need to introduce some form of marketing strategies like advertisements through newspapers and TV's. We need to make our product available globally at better rates and provide greater benefits then the competing company in efforts to cross a threshold of 50 percent.

We can flip the market by changing the parameter related to the strength of influence from global market to the local markets.
